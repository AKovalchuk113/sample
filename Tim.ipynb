```python
import pandas as pd
import random
from random import randint
```

```python
from datetime import date, datetime
```

```python
dates = []
values = []
i = 0
while i < 20:
    i+=1
    date1 = date(2023, 1, randint(1, 10))
    dates.append(date1)
    value = randint(30, 50)
    values.append(value)
```

```python
df = pd.DataFrame(columns = ['date', 'group', 'value'])
```

```python
df['date'] = dates
df['value'] = values
```

```python
groups = ['grp1', 'grp2', 'grp3']
for i in range(len(df)):
    df.at[i,'group'] = groups[randint(0,2)]
```

```python
#Само решение
res = []
for date2 in list(df['date'].unique()):
    buff = df[df['date'] == date2].reset_index(drop = True)
    summary = buff['value'].sum()
    date = date2
    for gr in list(df['group'].unique()):
        if len(buff[buff['group'] == gr]) > 0:
            val = buff[buff['group'] == gr]['value'].sum()
            fract = val/summary
            res.append([date2, gr, val, fract])
#         if len(buffer[buffer['group'] == gr]) == 0:
#             val = 0.0
#             fract = 0.0
#             res.append([date2, gr, val, fract])
# если хочешь видеть и нулевые на день то раскомменть
itog = pd.DataFrame(res, columns = ['date', 'group','value','fract'])
```
