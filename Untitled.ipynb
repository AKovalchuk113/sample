```python
#downloaded_symbols = pd.read_csv("companies.csv", sep=';')
companies = ['AMZN', 'TSLA', 'COST', 'GOOG', 'MSFT', 'NFLX', 'NVDA']
data = yf.download(companies, start="2015-01-01", end="2021-08-01", interval="1d")
df = data['Adj Close']
df = (df - df.shift(1)) / df.shift(1)
df.dropna(inplace = True)
df.plot(figsize=(16, 5), title='Diagramm')
#plt.semilogy()
#print(data)

mean_ = df.mean()
print("Вектор ожидаемых доходностей")
print(mean_)
cov_ = df.cov()
print("Ковариационная матрица доходностей")
print(cov_)
cov_inv = df_inv = pd.DataFrame(np.linalg.inv(cov_), cov_.columns, cov_.index)
print("Обратная к ковариационной матрице доходностей")
print(cov_inv)

tau = 0.6
#tau = 0.0

def func(x):
    return -2 * tau * np.dot(x, mean_) + np.dot(np.dot(x, cov_), x)


def portfolio_with_bounds():
    bounds = Bounds(np.zeros(len(mean_)), np.ones(len(mean_)))
    linear_constraint = LinearConstraint(np.ones(len(mean_)), [1], [1])
    x0 = np.array(np.zeros(len(mean_)))
    res = minimize(func, x0, method='trust-constr', constraints=[linear_constraint],
                   bounds=bounds, options={'xtol': 1e-30, 'gtol': 1e-30})
    return pd.Series(res.x, mean_.index)


def portfolio_without_bounds():
    alpha = np.dot(np.dot(np.ones(len(mean_)), cov_inv), np.ones(len(mean_)))
    beta = np.dot(np.dot(np.ones(len(mean_)), cov_inv), mean_)
    p = np.dot(cov_inv, np.ones(len(mean_))) / alpha + tau * (np.dot(cov_inv, mean_)
                                                              - beta / alpha * np.dot(cov_inv, np.ones(len(mean_))))
    return pd.Series(p, index=mean_.index)


def income(portfolio):
    data = yf.download(companies, start="2021-12-01", end="2021-12-01")
    open_date = data['Adj Close']
    data = yf.download(companies, start="2021-12-16", end="2021-12-16")
    close_date = data['Adj Close']
    capital = 10000
    sum_ = 0

    for i in range(len(portfolio)):
        sum_ += portfolio[i] / open_date.iloc[0][i] * close_date.iloc[0][i]
    return capital * sum_ - capital

#11.18
print('прибыль в долларах при тау=0: ' + str(0.005947*10000))
print('прибыль в долларах при тау=0.6: ' + str(0.01473*10000))
print('реальный доход (0): ' + str(0.005947*10000*1.1118))
print('реальный доход (0.6): ' + str(0.01473*10000*1.1118))

def answer(x):
    print("\n\nПортфель")
    print(x)
    print("\n\tОценка ожидаемой доходности")
    print(np.dot(x, mean_))
    print("\n\tОценка среднеквадратического отклонения доходности")
    print(np.sqrt(np.dot(np.dot(x, cov_), x)))


x1 = portfolio_with_bounds()
answer(x1)
```
